```python
#authorization
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api=tweepy.API(auth)
```

```python
#code twitter scraper 
number_of_tweets=10000

def search_tweets(query_string):
    with open(f"{query_string}.csv","w", newline='') as f:
        writer=csv.writer(f)
        writer.writerow(["tweet_id", "tweet_user_screen_name", "tweet_timestamp", "tweet_full_text", "tweet_retweet_count", "tweet_favorite_count","tweet.user.followers_count"])
        tweets=tweepy.Cursor(api.search_tweets, q=f"{query_string} -filter:retweets", tweet_mode="extended", result_type="recent", lang="en", count=100, until="2022-03-17").items(number_of_tweets) #since_id="1503520905674305548"
        for tweet in tweets:
            writer.writerow([tweet.id, tweet.user.screen_name, tweet.created_at, tweet.full_text.encode("utf-8"), tweet.retweet_count, tweet.favorite_count, tweet.user.followers_count])
```

## Word Cloud

```python
import numpy as np
import re
import emoji
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df2 = pd.read_csv('cryptopunks_combined.csv ', sep=';')
```

```python
def cleaner(tweet):
    tweet = re.sub("@[A-Za-z0-9]+","",tweet) #Remove @ sign
    tweet = re.sub(r"(?:\@|http?\://|https?\://|www)\S+", "", tweet) #Remove http links
    tweet = ''.join(c for c in tweet if c not in emoji.UNICODE_EMOJI) #Remove Emojis
    tweet = tweet.replace("#", "").replace("_", " ") #Remove hashtag sign but keep the text
    tweet = tweet.lower()
    tweet= re.sub('\xf0\x9f\x94\x9d+', " ", tweet)
    tweet= re.sub('x80', " ", tweet)
    #tweet= re.sub('b', " ", tweet)
    tweet= re.sub('xe2', " ", tweet)
    tweet= re.sub('x99t', " ", tweet)
    tweet= re.sub('x84', " ", tweet)
    tweet= re.sub('x8f', " ", tweet)
    tweet= re.sub('x99t', " ", tweet)
    tweet= re.sub('x9d', " ", tweet)
    tweet= re.sub('xb8', " ", tweet)
    tweet= re.sub('x9f', " ", tweet)
    tweet= re.sub('xf0', " ", tweet)
    tweet= re.sub('xa0', " ", tweet)
    tweet= re.sub('xc2', " ", tweet)
    tweet= re.sub('x83', " ", tweet)
    tweet= re.sub('xbb', " ", tweet)
    tweet= re.sub('xe3', " ", tweet)
    tweet= re.sub('x83', " ", tweet)
    tweet= re.sub("'", " ", tweet)
    tweet= re.sub("don", " ", tweet)
    tweet= re.sub("xef", " ", tweet)
    tweet= re.sub("x91", " ", tweet)
    tweet= re.sub("x87", " ", tweet)
    tweet= re.sub("xa8", " ", tweet)
    tweet= re.sub("x8d", " ", tweet)
    #tweet= re.sub("b", " ", tweet)
    tweet= re.sub("m003", " ", tweet)
    tweet= re.sub("x9a", " ", tweet)
    tweet= re.sub("x8d", " ", tweet)
    tweet= re.sub("xa6", " ", tweet)
    tweet= re.sub("nthis", " ", tweet)
    tweet= re.sub("\n", " ", tweet)
    #tweet= re.sub("`", " ", tweet)
    #tweet= re.sub('xa6', " ", tweet)
    tweet= re.sub(r'(?:^| )\w(?:$| )', ' ', tweet).strip()
    return tweet
df2['tweet_full_text'] = df2['tweet_full_text'].map(lambda x: cleaner(x))
```

```python
#just copy code for all the other dfs 

text=df2['tweet_full_text'].values 
wordcloud = WordCloud().generate(str(text))


# Display the generated image:
plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
