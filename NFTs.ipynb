# Start of KPI exploration

```python
#Nur den Zeilen-Count (die Länge des Dataframe) anzeigen 
print('Number of transfers:')
print(len(df.index))
```

```python
#Neuen df erstellen, der nur Nicht-Null-Werte für 'value' enthält

df_no_zeros=df[df.value != '0']

print('Anzahl Verkäufe:')
print(len(df_no_zeros))
```

```python
#Kumulation der value-Spalte mit sum()
#vorher to_numeric oder astype(float) konvertieren, da sonst die Zahlen als String aneinandergereiht werden

print('Total value traded in WEI:')
#print(pd.to_numeric(df_no_zeros['value'], errors='coerce').sum())
print(df_no_zeros['value'].astype(float).sum())

print('Total value traded in ETH:')
print((df_no_zeros['value'].astype(float).sum())/1000000000000000000)
```

```python
#Neuen df erstellen, der keine doppelten Werte für 'to_address' enthält

df_unique_buyers=df.drop_duplicates(subset=['to_address'])

print('Anzahl verschiedener Käufer:')
print(len(df_unique_buyers.index))
```

```python
#Neuen df erstellen, der keine doppelten Werte für 'from_address' enthält

df_unique_sellers=df.drop_duplicates(subset=['from_address'])

print('Anzahl verschiedener Verk:')
print(len(df_unique_sellers.index))
```
